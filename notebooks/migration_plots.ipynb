```python
s = mg.Sim('results3.hdf5')
s1 = mg.Sim('results2.hdf5')
cmass = np.cumsum(s.lam*s.dr[:,np.newaxis],axis=0)
cmass0 = np.cumsum(s.lamnp*s.dr[:,np.newaxis],axis=0)
mtot = cmass[-1,0]

cmass1 = np.cumsum(s1.lam*s1.dr[:,np.newaxis],axis=0)
cmass01 = np.cumsum(s1.lamnp*s1.dr[:,np.newaxis],axis=0)
mtot1 = cmass1[-1,0]
```

```python
skip = 5


fig,axes=plt.subplots(2,1,figsize=(10,5),sharex=True)
plt.subplots_adjust(hspace=0)

for i,a in enumerate(s.at[::skip]):
    axes[0].plot(s.rc/a,cmass[:,::skip][:,i]/(cmass0[:,::skip][:,i]),'-')
axes[0].plot(s.rc/s.at[-1],cmass[:,-1]/(cmass0[:,-1]),'--k',linewidth=2)

axes[0].axvline(1,color='k')
axes[0].axhline(1,color='k')
axes[0].set_ylim(.6,30)
axes[0].set_xscale('log')
axes[0].set_yscale('log')

for i,a in enumerate(s1.at[::skip]):
    axes[1].plot(s1.rc/a,cmass1[:,::skip][:,i]/(cmass01[:,::skip][:,i]),'-')
axes[1].plot(s1.rc/s1.at[-1],cmass1[:,-1]/(cmass01[:,-1]),'--k',linewidth=2)

axes[1].axvline(1,color='k')
axes[1].axhline(1,color='k')
axes[1].set_ylim(.6,30)
axes[1].set_xscale('log')
axes[1].set_yscale('log')

axes[1].set_xlabel('$r/a$',fontsize=20)
axes[0].set_ylabel('$M_<(r)$',fontsize=20)
axes[1].set_ylabel('$M_<(r)$',fontsize=20)
axes[0].text(.01,10,'Fast',fontsize=20)
axes[1].text(.01,10,'Slow',fontsize=20)

fig.savefig('cumulative_mass.pdf')
```
