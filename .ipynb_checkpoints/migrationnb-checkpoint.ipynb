```python
%matplotlib inline
import matplotlib.pyplot as plt
import mpld3
import numpy as np
import scipy as sp
import scipy.optimize as opt
import scipy.integrate as integ
mpld3.enable_notebook()
```

```python
class Params():
    def __init__(self,**kargs):
        self.a=kargs['a']
        self.gam = kargs['gamma']
        self.mu = kargs['mu']
        self.h = kargs['h']
        self.alpha = kargs['alpha']
        self.beta = kargs['beta']
        self.delta = kargs['delta']
        self.G1 = kargs['G1']
        self.ri = kargs['ri']
        self.ro = kargs['ro']
        self.lamo = kargs['lamo']
        self.lami = kargs['lami']
        self.nr = kargs['nr']
        self.mdot0 = 1.5*(self.lamo-self.lami)/(self.ro/self.nu(self.ro)-self.ri/self.nu(self.ri))
        self.mth = self.h**3
        self.rh = self.a*(self.mu*self.mth/3)**(1./3)
        self.dep = max(self.h,self.rh)
        self.T0 = 2*np.pi*self.a*self.mu**2 * self.h**2
        self.r = np.linspace(self.ri,self.ro,self.nr)
        if kargs['planet_torque']:
            self.planet_torque = 1
        else:
            self.planet_torque = 0
            
    def nu(self,r):
        return self.alpha*self.h**2 * r**(self.gam)
    def adisk(self,r):
        return self.lami + (2*self.mdot0/3)*(r/self.nu(r)-self.ri/self.nu(self.ri))
    def Tfunc(self,r):
        x = (r-self.a)/(self.rh)
        fac1 = (x-self.beta)/self.delta
        fac2 = (x + self.beta )/self.delta
        norm = np.sqrt(np.pi)*self.delta
        res=(self.G1+1)*np.exp(-fac1**2)-np.exp(-fac2**2)
        return self.T0*res/norm
    def vr0(self,r):
        return -self.mdot0/self.adisk(r)
    
    def derivs(self,x,y):
        mdot = y[1]
        lam = y[0]
        nuval = 3*self.nu(x)
        fac1 = mdot/nuval
        fac2 = - lam*((self.gam-.5)/x  - self.planet_torque*self.Tfunc(x)/(nuval*np.pi*np.sqrt(x)))
        
        return [fac1+fac2,0]
                               
    def jac(self,x,y):
        nuval = 3*self.nu(x)
        bx = 1./(nuval)
        ax = -((self.gam-.5)/x - self.planet_torque*self.Tfunc(x)/(nuval*np.pi*np.sqrt(x)))
        return [[ax,bx],[0,0]]
    
    def integrate(self,m0):
        y0 = [self.lamo,m0]
        sys = integ.ode(self.derivs, self.jac).set_integrator('lsoda',with_jacobian=True)
        sys.set_initial_value(y0, self.ro)
        i = len(self.r)-2
        lam = [y0[0]]
        while sys.successful() and sys.t > self.ri:
            try:
                sys.integrate(self.r[i])
            except: 
                print 'Error: not converging'
                return [-1]
            lam.append(sys.y[0])
            i -= 1
        return np.array(lam)[::-1]
    def func(self,m):
        res = self.integrate(m)[0]
        if res == -1:
            return 0
        if self.lami == 0:
            return res
        else:
            return res/self.lami - 1
        
    def shoot(self):
        mres = opt.newton(self.func,self.mdot0)
        err = self.integrate(mres)[0]-self.lami
        if  err == -1 or np.abs(err)>1e10:
            print 'Error above 10^10, error = ', err
            return self.mdot0
        return mres
    def steady_state(self):
     
        mdot = self.shoot()
        if mdot == self.mdot0:
            lam = self.adisk(self.r)
        else:
            lam = self.integrate(mdot)
        
        
        self.mdot_ss = mdot
        self.lam_ss = lam
        self.vr_ss = -mdot/lam
        return mdot,lam
        
pargs={ 'a': 10,
        'gamma' :.5,
        'mu' : 1,
        'h': .1,
        'alpha': .06,
        'beta' : 2./3,
        'delta' : 1,
        'G1': 0,
        'ri': 1,
        'ro': 100,
        'lami': 1e-3,
        'lamo': 1,
       'nr': 1e3,
      'planet_torque': True}    

pars = Params(**pargs)
```

```python

pars.alpha = .01
md,lam=pars.steady_state()

fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].loglog(pars.r,pars.lam_ss,'-r')
ax[0].loglog(pars.r,pars.adisk(pars.r),'--b')
ax[0].set_title('$\\dot{M} = %.2e$' % md,fontsize=20)
ax[0].set_xlabel('$r$',fontsize=20)
ax[0].set_ylabel('$\\lambda$',fontsize=20)
ax[1].loglog(pars.r,pars.vr_ss/pars.vr0(pars.r),'-r')
ax[1].axhline(1)
ax[1].set_xlabel('$r$',fontsize=20)
ax[1].set_ylabel('$\\frac{v_r}{v_r^0}$',fontsize=20)
```

```python
alist= np.array([.01,.03,.06,.1,.3,.6,1])

sim = []
for a in alist:
    pargs['alpha'] = a
    temp = Params(**pargs)
    md,lam = temp.steady_state()
    sim.append(temp)

fig,ax = plt.subplots(2,2,figsize=(10,5))
for s in sim:
    ax[0,0].loglog(s.r,s.lam_ss)
    ax[0,0].loglog(s.r,s.adisk(s.r),'--')
    ax[0,0].set_title('$\\dot{M} = %.2e$' % s.mdot0,fontsize=20)
    ax[0,0].set_xlabel('$r$',fontsize=20)
    ax[0,0].set_ylabel('$\\lambda$',fontsize=20)
    ax[0,1].loglog(s.r,s.vr_ss/s.vr0(s.r))
    ax[0,1].axhline(1)
    ax[0,1].set_xlabel('$r$',fontsize=20)
    ax[0,1].set_ylabel('$\\frac{v_r}{v_r^0}$',fontsize=20)
ax[1,0].loglog(alist,[s.mdot_ss for s in sim],'-ko')
ax[1,0].axhline(sim[0].mdot0)
ax[1,0].set_xlabel('$\\alpha$',fontsize=20)
ax[1,0].set_ylabel('$\\dot{M}$',fontsize=20)
```

```python
lolist= np.array([.003,.01,.03,.1,.3,1])

sim = []
for l in lolist:
    pargs['lamo'] = l
    temp = Params(**pargs)
    md,lam = temp.steady_state()
    sim.append(temp)

fig,ax = plt.subplots(2,2,figsize=(10,5));
for s in sim:
    ax[0,0].loglog(s.r,s.lam_ss)
    ax[0,0].loglog(s.r,s.adisk(s.r),'--')
    ax[0,0].set_title('$\\dot{M} = %.2e$' % s.mdot0,fontsize=20)
    ax[0,0].set_xlabel('$r$',fontsize=20)
    ax[0,0].set_ylabel('$\\lambda$',fontsize=20)
    ax[0,1].semilogx(s.r,s.vr_ss/s.vr0(s.r))
    ax[0,1].axhline(1)
    ax[0,1].set_xlabel('$r$',fontsize=20)
    ax[0,1].set_ylabel('$\\frac{v_r}{v_r^0}$',fontsize=20)
ax[1,0].loglog(lolist,[s.mdot_ss for s in sim],'-ko')
ax[1,0].axhline(sim[0].mdot0)
ax[1,0].set_xlabel('$\\lambda_o$',fontsize=20)
ax[1,0].set_ylabel('$\\dot{M}$',fontsize=20)
```
